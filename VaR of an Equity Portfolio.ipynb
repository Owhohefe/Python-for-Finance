```python
import numpy as np
import pandas as pd
import math
from scipy.stats import norm
```

```python
mydata = pd.read_excel(r"C:\Users\oekpom\Desktop\SOP\price_list.xlsx")
```

```python
mydata.head()
```

```python
mydata = mydata.set_index("Date")
```

```python
mydata.head()
```

```python
returns = np.log(mydata/mydata.shift(1))
returns
```

```python
weights = np.array([0,0.041113002,0.198186371,0.001455058,0.096273662,0.224522367,0,0.000200941,0.014039035,0,0.0000788219543874859,0.328176956,0,0,0,0.095953786,0,0,0,0,0])
```

```python
cov_matrix = returns.cov()
cov_matrix
```

```python
FBNH_var = returns["FBNH"].var()
FBNH_var
```

```python
portfolio_variance = np.dot(weights.T, np.dot(returns.cov(), weights))
portfolio_variance
```

```python
portfolio_volatility = np.dot(weights.T, np.dot(returns.cov(), weights))**0.5
portfolio_volatility
```

```python
print(str(round(portfolio_volatility,5)*100) +"%")
```

```python
Market_Value = 35639816.62
Confidence_Level = norm.ppf(0.99)
Holding_period = math.sqrt(10)
```

```python
VaR_10_Days = Market_Value*portfolio_volatility*Confidence_Level*Holding_period
VaR_10_Days
```
