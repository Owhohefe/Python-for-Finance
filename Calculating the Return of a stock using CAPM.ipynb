```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
```

```python
tickers = ["PG", "^GSPC"]
mydata = pd.DataFrame()
for t in tickers:
    mydata[t]=wb.DataReader(t, data_source = "yahoo", start="2012-1-1", end="2016-12-31")["Adj Close"]
```

```python
mydata.head()
```

```python
returns = np.log(mydata/mydata.shift(1))
returns
```

```python
cov = returns.cov() * 250
cov
```

```python
cov_with_market = cov.iloc[0,1]
cov_with_market
```

```python
market_var = returns["^GSPC"].var() * 250
market_var
```

```python
PG_beta = cov_with_market/market_var
PG_beta
```

```python
market_return_annual = returns["^GSPC"].mean() * 250
market_return_annual
```

```python
risk_free_rate = 0.025 #10 year bond yield
PG_Expected_return = risk_free_rate + PG_beta * (market_return_annual - risk_free_rate)
PG_Expected_return
```

```python
print(str(round(PG_Expected_return,5)*100) +"%")
```
